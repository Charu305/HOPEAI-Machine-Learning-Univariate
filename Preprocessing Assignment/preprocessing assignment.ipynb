```python
dataset = pd.read_csv("kidney_disease.csv")
```

```python
dataset.isnull().sum()
```

```python
import numpy as np

dataset = dataset.replace([
    r'^\s*\\t\?\s*$',   # literal "\t?" 
    r'^\s*\?\s*$',      # just "?"
    r'^\s*\t.*$',       # values starting with an actual tab character
], np.nan, regex=True)
```

```python
dataset[["pcv", "wc", "rc"]] = dataset[["pcv", "wc", "rc"]].astype(float)
```

```python
def quanQual(dataset):
    quan=[]
    qual=[]
    for columnName in dataset.columns:
        #print(columnName)
        if(dataset[columnName].dtype=='O'):
            #print("qual")
            qual.append(columnName)
        else:
            #print("quan")
            quan.append(columnName)
    return qual,quan
```

```python
qual,quan = quanQual(dataset)
```

```python
qual
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(dataset[quan])
dataset[quan]=imp.transform(dataset[quan])
```

```python
from sklearn.impute import SimpleImputer

imp_cat = SimpleImputer(strategy='most_frequent')
dataset[qual] = imp_cat.fit_transform(dataset[qual])
```
